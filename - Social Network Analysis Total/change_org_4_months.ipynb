```python
import os
```

```python
csv_files = []
for file in os.listdir():
    if file.split(".")[-1] == "csv":
        csv_files.append(file)
```

```python
csv_files
```

```python
len(csv_files)
```

```python
import pandas as pd
```

```python
all_petitions_comments_1 = pd.read_csv(csv_files[0])
all_petitions_comments_2 = pd.read_csv(csv_files[1])
```

```python
all_petitions_comments_1.head(1)
```

```python
all_petitions_comments_2.head(1)
```

```python
all_petitions_comments_1["created_at"] = pd.to_datetime(all_petitions_comments_1["created_at"])
```

```python
all_petitions_comments_2["created_at"] = pd.to_datetime(all_petitions_comments_2["created_at"])
```

```python
all_petitions_comments = pd.concat([all_petitions_comments_1, all_petitions_comments_2])
```

```python
all_petitions_comments_4_months = all_petitions_comments[
    (all_petitions_comments.created_at.astype(str) >= "2021-01-01 00:00:00")
    & (all_petitions_comments.created_at.astype(str) < "2021-05-01 00:00:00")
]
```

```python
all_petitions_comments_4_months = all_petitions_comments_4_months.reset_index().drop(columns=["level_0", "index"])
```

```python
all_petitions_comments_4_months.to_csv("4_months_data/all_petitions_comments_4_months.csv", index=False)
```

```python
all_petitions_comments_4_months.to_json("4_months_data/all_petitions_comments_4_months.json", orient="records")
```
